```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

* Calculate the average reading score

```python
#Calculate the total number of schools
total_num_school = school_df['school_name'].nunique()
# #Calculate the total number of students
total_num_stu = school_df['Student ID'].nunique()
# #Calculate the total budget
total_budget = school_df['budget'].unique().sum()
# #Calculate the average math score#
ave_math = school_df['math_score'].mean()
# #Calculate the average reading score
ave_reading = school_df['reading_score'].mean()
# # Calculate the percentage of students with a passing math score (70 or greater)
passing_math=round((len(school_df[school_df['math_score'] >= 70])/total_num_stu)*100, 6)
#Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading=round((len(school_df[school_df['reading_score'] >= 70])/total_num_stu)*100, 6)
#Calculate the percentage of students who passed math **and** reading (% Overall Passing)
overall_passing =((school_df[(school_df['math_score']>=70)&
                              (school_df['reading_score']>=70)].count()['student_name'])/total_num_stu)*100
#Create a dataframe to hold the above results
summery_df= pd.DataFrame([[total_num_school, total_num_stu, total_budget, ave_math, 
                           ave_reading, passing_math, passing_reading, overall_passing]])
DistrictSummary_df= pd.DataFrame({
    "total schools": [total_num_school],
    "total students":[total_num_stu],
    "total budget": [total_budget],
    "Average Math Score": [ave_math],
    "Average Reading Score": [ave_reading],
    "Passing Math": [passing_math],
    "Passing Reading": [passing_reading],
    "Overall Passing": [overall_passing]
})
DistrictSummary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#School Name
school_names = school_df.set_index("school_name").groupby(["school_name"])
#School Type
school_type = school_data.set_index("school_name")["type"]
#Total Students
total_students = school_names["Student ID"].count()
#Total School Budget
total_school_budget = school_data.set_index("school_name")["budget"]
#Per Student Budget
budget_per_student = (total_school_budget)/(total_students)
#Average Math Score
average_math_score = school_names["math_score"].mean()
#Average Reading Score
average_reading_score = school_names["reading_score"].mean()
# % Passing Math
percent_passing_math = (school_df[school_df["math_score"] >= 70].groupby("school_name")["Student ID"].count()/total_students)*100
# % Passing Reading
percent_passing_reading = (school_df[school_df["reading_score"] >= 70].groupby("school_name")["Student ID"].count()/total_students)*100
# % Overall Passing
percent_overall_passing = (school_df[(school_df["math_score"] >= 70) & 
                                    (school_df["reading_score"] >= 70)].groupby("school_name")["Student ID"].count()/total_students) *100
#Create a dataframe to hold the above results
school_summary_df = pd.DataFrame({
   "School Type": school_type,
   "Total Students": total_students,
   "Total School Budget": total_school_budget,
    "Per Student Budget": budget_per_student,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "% Passing Math": percent_passing_math,
    "% Passing Reading": percent_passing_reading,
    "% Overall Passing": percent_overall_passing})

school_summary_df

```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Set bins 
bins = [0, 585, 630, 645, 680]
bin_labels = ["<585", "585-630", "630-645", "645-680"]

school_df["bins"] = pd.cut(school_df["budget"]/ school_df["size"], bins , labels = bin_labels)

#setting table
bins_groupby = school_df.groupby("bins")

#Average Math Score
average_math_score_x = bins_groupby["math_score"].mean()
#Average Reading Score
average_reading_score_x = bins_groupby["reading_score"].mean()
# % Passing Math
percent_passing_math_x = (school_df[school_df["math_score"] >= 70].groupby("bins")["Student ID"].count()/bins_groupby["Student ID"].count())*100
# % Passing Reading
percent_passing_reading_x = (school_df[school_df["reading_score"] >= 70].groupby("bins")["Student ID"].count()/bins_groupby["Student ID"].count())*100
#Overall Passing Rate
overall_passing_rate_x = (school_df[(school_df["reading_score"] >= 70) & (school_df["math_score"] >= 70)].groupby('bins')["Student ID"].count()/bins_groupby["Student ID"].count())*100

#data frame

scores_by_school_spending_df = pd.DataFrame({
    "Average Math Score":average_math_score_x,
    "Average Reading Score":average_reading_score_x,
    "% Passing Math":percent_passing_math_x,
    "% Passing Reading":percent_passing_reading_x,
    "Overall Passing Rate":overall_passing_rate_x
})

scores_by_school_spending_df
                             
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
#set bins
bins_y = [0,1000,2000,5000]
bin_labels_y = ["Small (<1000)" , "Medium (1000-2000)", "Large (2000-5000)"]

school_df["bins_y"] = pd.cut(school_df["size"], bins_y , labels = bin_labels_y)

#setting table
bins_groupby_y = school_df.groupby("bins_y")

#Average Math Score
average_math_score_y = bins_groupby_y["math_score"].mean()
#Average Reading Score
average_reading_score_y = bins_groupby_y["reading_score"].mean()
# % Passing Math
percent_passing_math_y = (school_df[school_df["math_score"] >= 70].groupby("bins_y")["Student ID"].count()/bins_groupby_y["Student ID"].count()) *100
# % Passing Reading
percent_passing_reading_y = (school_df[school_df["reading_score"] >= 70].groupby("bins_y")["Student ID"].count()/bins_groupby_y["Student ID"].count())*100
#Overall Passing Rate
overall_passing_rate_y = (school_df[(school_df["reading_score"] >= 70) & (school_df["math_score"] >= 70)].groupby('bins_y')["Student ID"].count()/bins_groupby_y["Student ID"].count())*100

#data frame

scores_by_school_size_df = pd.DataFrame({
    "Average Math Score":average_math_score_y,
    "Average Reading Score":average_reading_score_y,
    "% Passing Math":percent_passing_math_y,
    "% Passing Reading":percent_passing_reading_y,
    "Overall Passing Rate":overall_passing_rate_y
})

scores_by_school_size_df
```
